```python
# While in csv format, accessed csv file to format 'Value' column into number format.  This was necessary because the original format was a string with periods in place of decimals that could not be removed without changing the underlying number values.
usda_clean_df=pd.read_excel(r'C:\Users\watso\Documents\Data_Science_Bootcamp\exploratory_data_analysis_project\usda_df_clean.xlsx')
```

```python
#created a dataframe using the cleaned up file with the properly formatted 'Value' column
usda_clean_mod_df=usda_clean_df[['group_desc','reference_period_desc', 'load_time', 'country_name', 'source_desc', 'country_code', 'location_desc', 'state_alpha', 'prodn_practice_desc', 'commodity_desc', 'sector_desc', 'agg_level_desc', 'class_desc', 'year',  'state_name', 'Value', 'short_desc', 'domaincat_desc', 'state_ansi', 'domain_desc', 'statisticcat_desc', 'util_practice_desc', 'freq_desc']]
usda_clean_mod_df
```

```python
# Dataframe containing only those rows for the year 2020 where the column 'short_desc' indicates that the number in 'Value' indicates number of colonies lost through colony collapse disorder.
usda_collapse_df = usda_clean_mod_df[usda_clean_mod_df['short_desc'].str.contains("COLONY COLLAPSE DISORDER")]
usda_collapse_df1 =usda_collapse_df[usda_collapse_df['year']==2020]
usda_collapse_df1
```

```python
# Dataframe containing only those rows for the year 2020 where the column 'short_desc' indicates that the number in 'Value' indicates a percentage of colonies affected by pesticides.
usda_pesticide_df = usda_clean_mod_df[usda_clean_mod_df['short_desc'].str.contains("AFFECTED BY PESTICIDES")]
usda_pesticide_df1 = usda_pesticide_df[usda_pesticide_df['year']==2020]
usda_pesticide_df1
```

```python
# Dataframe containing only those rows for the year 2020 where the column 'short_desc' indicates that the number in 'Value' indicates the max inventory total of colonies by time period. This will be used to convert the 'Value' column for the pesticide data frame into a usable number of colonies integer amount.
usda_inventory_df = usda_clean_mod_df[usda_clean_mod_df['short_desc'].str.contains("INVENTORY, MAX,")]
usda_inventory_df1 = usda_inventory_df[usda_inventory_df['year']==2020]
usda_inventory_df1
```
